```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("database/FIIS23.s3db")
cur = conn.cursor()
cur.execute("SELECT * FROM TablaFIIS")
filas = cur.fetchall()
conn.commit()
```

```python
print ("Weeks\t|Disponibilidad de la Flota\t|Vehiculos que regresan de mantenimiento\t|Vehiculos que se retiran de mantenimiento\t|Vehiculos asignados a operar")
print ("----------------------------------------------------------------------------------------------------")
print ("")

for i in filas:
    print(i[0], "\t|", i[3], "\t|", i[12], "\t|", i[13], "\t|", i[10])
```

```python
print ("Weeks\t|Cantidad de vehiculos en operaci√≥n\t|Vehiculos asignados a operar\t|Vehiculos devueltos al deposito")
print ("----------------------------------------------------------------------------------------------------")
print ("")

for i in filas:
    print(i[0], "\t|", i[2], "\t|", i[10], "\t|", i[11])
```

```python
print ("Weeks\t|Cantidad de fallas mecanicas\t|Numero de veces que ocurre una falla mecanica\t|Reparacion de fallas mecanicas\t|Cantidad de reparaciones preventivas")
print ("----------------------------------------------------------------------------------------------------")
print ("")

for i in filas:
    print(i[0], "\t|", i[1], "\t|", i[8], "\t|", i[9], "\t|", i[7])
```

```python
print ("Weeks\t|Disponibilidad de talleres\t|Cantidad de plazas sin asignacion de mantenimiento\t|Cantidad de plazas apartadas por mantenimiento")
print ("----------------------------------------------------------------------------------------------------")
print ("")

for i in filas:
    print(i[0], "\t|", i[4], "\t|", i[6], "\t|", i[5])
```

```python
df = pd.read_sql_query("SELECT * FROM TablaFIIS", conn)
conn.close()
```

```python
df
```

```python
variables_y = [
    "disponibilidadDeLaFlota",
    "vehiculosQueRegresanDeMantenimiento",
    "vehiculosQueSeRetiranPorMantenimiento",
    "vehiculosAsignadosAOperar"
]

plt.figure()

for variable in variables_y:
    plt.plot(df["sem"], df[variable], label=variable)

plt.xlabel("Semana")
plt.ylabel("Valor")
plt.title("Nivel: Disponibilidad de la flota")
plt.legend(loc="best", bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
variables_y = [
    "cantidadDeVehiculosEnOperacion",
    "vehiculosAsignadosAOperar",
    "vehiculosDevueltosAlDeposito"
]

plt.figure()

for variable in variables_y:
    plt.plot(df["sem"], df[variable], label=variable)

plt.xlabel("Semana")
plt.ylabel("Valor")
plt.title("Nivel: Disponibilidad de la flota")
plt.legend(loc="best", bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots()

# Primer eje Y (izquierda)
ax1.plot(df["sem"], df["disponibilidadDeTalleres"], label="disponibilidadDeTalleres", color="blue")
ax1.set_xlabel("Semana")
ax1.set_ylabel("Disponibilidad de Talleres", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# Segundo eje Y (derecha)
ax2 = ax1.twinx()
ax2.plot(df["sem"], df["cantidadDePlazasSin"], label="cantidadDePlazasSin", color="orange")
ax2.plot(df["sem"], df["cantidadDePlazasApartadasPorMantenimiento"], label="cantidadDePlazasApartadasPorMantenimiento", color="green")
ax2.set_ylabel("Cantidad de Plazas", color="green")
ax2.tick_params(axis='y', labelcolor="green")

# T√≠tulo y rejilla
plt.title("Nivel: Disponibilidad de la flota")
fig.tight_layout()
plt.grid(True)
plt.show()
```

```python
variables_y = [
    "disponibilidadDeLaFlota",
    "vehiculosQueRegresanDeMantenimiento",
    "vehiculosQueSeRetiranPorMantenimiento",
    "vehiculosAsignadosAOperar"
]
fig, axs = plt.subplots(nrows=4, sharex=True, figsize=(10, 8))
colores = ['blue', 'orange', 'green', 'red']
for i, var in enumerate(variables_y):
    axs[i].plot(df["sem"], df[var], color=colores[i])
    axs[i].set_ylabel("") 
    axs[i].grid(True)
plt.xlabel("Semana")
plt.suptitle("Nivel: Disponibilidad de la flota", fontsize=14)
fig.legend(variables_y, loc="upper left", bbox_to_anchor=(1.0, 0.5))
plt.tight_layout(rect=[0, 0.03, 0.85, 0.95])
plt.show()
```
